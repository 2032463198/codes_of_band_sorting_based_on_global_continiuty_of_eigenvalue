```python
import numpy as np

deltak0=1/60            #origial mesh fineness
a=2e-7                  #period along x direction
b=2e-7                  #period along y direction
filetitle="this_example-IBZ"     #filetitle should illustrate the character of a sorting task

kx=[]
ky=[]
size=int(1/deltak0)
for i in range(0,size+1):
    for j in range(i,size+1):
        kx.append(j*deltak0*np.pi/a)
        ky.append(i*deltak0*np.pi/b)
np.savetxt('kpoints_x_%s.txt'%(filetitle), np.array(kx), newline=' ', fmt='%s')
np.savetxt('kpoints_y_%s.txt'%(filetitle), np.array(ky), newline=' ', fmt='%s')
np.savetxt('basic_info_%s.txt'%(filetitle), np.array([len(ky),deltak0,a,b]), newline=' ', fmt='%s')
print("Number of original k pionts: %d"%(len(kx)))
```
